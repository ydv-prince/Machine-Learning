```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

from sklearn.neighbors import KNeighborsClassifier  #KNN
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_classification(n_samples= 1000, n_features= 5, n_classes= 2, n_redundant= 0, random_state= 32)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 32)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf = KNeighborsClassifier()
clf
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Evaluate model

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print("Classification Report")
print(classification_report(y_test, y_pred))
print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
params = {
    'n_neighbors': [3,5,6,7,9,11,13],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [20, 30, 40, 50]
}
```

```python
grid = GridSearchCV(estimator= KNeighborsClassifier(), param_grid= params, cv= 5, verbose= 3)
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned
```

```python
# Evaluate tuned model

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_tuned)}")
print("Classification Report")
print(classification_report(y_test, y_pred_tuned))
print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred_tuned))
```
