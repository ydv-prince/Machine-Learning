```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_regression(n_samples= 1000, n_features= 2, noise= 7, random_state= 25)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 25)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = KNeighborsRegressor()
reg
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
#Evaluate

print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
# hyperparameter tuning
params = {
    'n_neighbors': [3,5,2,8,5,9,11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'kd_tree', 'ball_tree', 'brute'],
    'leaf_size': [20, 30, 40, 50]
}

params
```

```python
grid = GridSearchCV(estimator= KNeighborsRegressor(), param_grid= params, verbose= 3, cv = 5)
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_tuned_pred = best_model.predict(X_test)
```

```python
# Evaluate tuned model

print(f"R2 Score: {r2_score(y_test, y_tuned_pred)}")
```
