```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_regression(n_samples= 1000, n_features= 20, noise= 0.1, random_state= 1) #random_state= 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state= 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor()
svr = SVR(kernel= 'linear')
```

```python
# to create ensemble

from sklearn.ensemble import VotingRegressor
ensemble_regressor = VotingRegressor(estimators= [("logistic_regression", lr), ("decision_tree_regressor", dtr), ("support_vector_regressor", svr)])
ensemble_regressor
```

```python
ensemble_regressor.fit(X_train, y_train)
```

```python
y_pred = ensemble_regressor.predict(X_test) #Average of all the prediction will be the final prediction
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
