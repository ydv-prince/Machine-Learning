```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples= 1000, n_features= 20, n_classes= 2, random_state= 45) #random_state= 1
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state= 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
lr_clf = LogisticRegression()
dt_clf = DecisionTreeClassifier()
sv_clf = SVC()
gn_clf = GaussianNB()
```

```python
# for ensembles

from sklearn.ensemble import VotingClassifier
```

```python
ensemble_clf = VotingClassifier(estimators= [("Logistic_regression_clf", lr_clf), ("decision_tree_clf", dt_clf), ("support_vector_clf", sv_clf), ("naive_bayes_clf", gn_clf)], voting= "hard")
```

```python
ensemble_clf
```

```python
ensemble_clf.fit(X_train, y_train)
```

```python
y_pred = ensemble_clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
