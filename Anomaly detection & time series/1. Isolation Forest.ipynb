```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
out = IsolationForest(contamination= 0.2)
out
```

```python
out.fit(df)
```

```python
pred = out.predict(df)
pred
```

```python
index = np.where(pred<0)
index
```
