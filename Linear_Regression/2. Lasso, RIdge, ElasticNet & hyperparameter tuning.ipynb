```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop("name", axis=1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['origin'].value_counts()
```

```python
#data encoding
df["origin"] = df["origin"].map({"usa":1, "japan":2, "europe":3})
```

```python
df.info()
```

```python
X = df.drop('mpg', axis= 1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
#train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
# CV with hyperparameter tuning

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# gridsearchCV for lasso
# define by model

lasso = Lasso()
```

```python
lasso
```

```python
#grid search cv>> Dictionary with parameters names (str) as keys and lists of parameter settings to try as value

param_grid = {'alpha': [0.001, 0.01, 0.1, 1,2,5,10,100]}
param_grid
```

```python
grid_search = GridSearchCV(estimator= lasso, param_grid= param_grid, cv= 5, scoring= 'r2', verbose= 2)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_
#Mean cross-validated score of the best_estimator
```

```python
grid_search.best_estimator_
#Estimator that was chosen by the search, i.e. estimator which gave highest score (or smallest loss if specified) on the left out data
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
r2_grid = r2_score(y_test, y_pred)
print(f"The R square of GridSearchCV of Lasso regression: {r2_grid}")
```

```python
# randomized search cv for lasso

lasso = Lasso()
param_distributions = {'alpha': [0.01,0.1,1,2,10,100]}
random_search_lasso = RandomizedSearchCV(estimator= lasso, param_distributions= param_distributions, cv= 5, scoring='r2', verbose= 2, n_iter= 3)
random_search_lasso
```

```python
random_search_lasso.fit(X_train, y_train)
```

```python
random_search_lasso.best_estimator_
```

```python
random_search_lasso.best_score_
```

```python
y_random_pred = random_search_lasso.predict(X_test)
r2_random_lasso = r2_score(y_test, y_pred)
print(f"The R square of RandomizedSearchCV of Lasso regression: {r2_random_lasso}")
```

```python
# grid search cv for ridge

ridge = Ridge()
param_grid = {'alpha': [0.01,0.1,1,2,10,100]}
grid_search_ridge = GridSearchCV(estimator= ridge, param_grid= param_grid, cv= 5, scoring='r2', verbose= 2)
grid_search_ridge.fit(X_train, y_train)
print("\n")
print(grid_search_ridge.best_estimator_)
print(grid_search_ridge.best_score_)
y_grid_pred = grid_search_ridge.best_estimator_.predict(X_test)
r2_random_ridge = r2_score(y_test, y_pred)
print(f"The R square of GridSearchCV of Ridge regression: {r2_random_ridge}")
```

```python
# randomized search cv for ridge

ridge = Ridge()
param_distributions = {'alpha': [0.01,0.1,1,2,10,100]}
random_search_ridge = RandomizedSearchCV(estimator= ridge, param_distributions= param_distributions, cv= 5, scoring='r2', verbose= 2, n_iter= 4)
random_search_ridge.fit(X_train, y_train)
print("\n")
print(random_search_ridge.best_estimator_)
print(random_search_ridge.best_score_)
y_random_pred = random_search_ridge.predict(X_test)
r2_random_ridge = r2_score(y_test, y_pred)
print(f"The R square of RandomizedSearchCV of Ridge regression: {r2_random_ridge}")
```

```python
# grid search cv for ElasticNet

model = ElasticNet()
param_grid = {'alpha': [0.01,0.1,1,2,10,100],
              'l1_ratio': [0.1,0.4,0.9]}
grid_search_elastic = GridSearchCV(estimator= model, param_grid= param_grid, cv= 5, scoring='r2', verbose= 2)
grid_search_elastic.fit(X_train, y_train)
print("\n")
print(grid_search_elastic.best_estimator_)
print(grid_search_elastic.best_score_)
y_grid_pred = grid_search_elastic.best_estimator_.predict(X_test)
r2_grid_elastic = r2_score(y_test, y_grid_pred)
print(f"The R square of GridSearchCV of ElasticNet regression: {r2_grid_elastic}")
```

```python
# randomized search cv for ElasticNet

model = ElasticNet()
param_distributions = {'alpha': [0.01,0.1,1,2,5,10,100],
              'l1_ratio': [0.1,0.4,0.9]}
random_search_elastic = RandomizedSearchCV(estimator= model, param_distributions= param_distributions, n_iter= 4, cv= 5, scoring='r2', verbose= 2)
random_search_elastic.fit(X_train, y_train)
print("\n")
print(random_search_elastic.best_estimator_)
print(random_search_elastic.best_score_)
y_grid_pred = random_search_elastic.best_estimator_.predict(X_test)
r2_random_elastic = r2_score(y_test, y_grid_pred)
print(f"The R square of RandomizedSeaarchCV of Elastic regression: {r2_random_elastic}")
```
