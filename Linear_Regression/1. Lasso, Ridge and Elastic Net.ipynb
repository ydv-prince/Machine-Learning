```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
df.drop("name", axis=1, inplace = True)
```

```python
df
```

```python
df.isna().sum()
```

```python
# As we fhave not done outlier treatment, so better to repace the missing with median
```

```python
df['horsepower'].median()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['origin'].value_counts()
```

```python
#convert string column to numerical

df['origin'] = df['origin'].map({"usa": 1, "japan":2, "europe":3})
```

```python
df['origin'] = df['origin'].astype(int)
```

```python
#separate X and y
X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#multiple linear regression model
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficients for {col_name} is {regression_model.coef_[i]}")
```

```python
# Observations: Coefficients are relatively smaller, if one independent variable changes
# there will be not much difference in prediction this is sometime called as smoother model

# these features moght not be contributing in the model training
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"The R square of linear regression: {r2_linear}")
```

```python
from sklearn.linear_model import Ridge
# In practical implementation lambda is alpha

ridge_regression_model = Ridge(alpha=0.1)
```

```python
ridge_regression_model
```

```python
ridge_regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficients for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
# For ridge regression evaluation

y_pred_ridge = ridge_regression_model.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"The R square of Ridge regression: {r2_ridge}")
```

```python
# We don't see much in coefficient as compared to linear regression
```

```python
from sklearn.linear_model import Lasso
lasso_regression_model = Lasso(alpha= 0.5)
```

```python
lasso_regression_model
```

```python
lasso_regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficients for {col_name} is {lasso_regression_model.coef_[i]}")
```

```python
# Three features coefficients is 0, lasso helps in feature selection
```

```python
y_pred_lasso = lasso_regression_model.predict(X_test)
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f"The R square of Lasso regression: {r2_lasso}")
```

```python
from sklearn.linear_model import ElasticNet
elastic_net_model = ElasticNet(alpha= 1, l1_ratio= 0.5)
```

```python
elastic_net_model
```

```python
elastic_net_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficients for {col_name} is {elastic_net_model.coef_[i]}")
```

```python
y_pred_elastic = elastic_net_model.predict(X_test)
r2_elastic = r2_score(y_test, y_pred_elastic)
print(f"The R square of Elastic regression: {r2_elastic}")
```

```python
#regularisation with crossvalidation

from sklearn.linear_model import LassoCV
lasso_cv = LassoCV(cv= 5, verbose= 1)
```

```python
lasso_cv
```

```python
lasso_cv.fit(X_train,y_train)
```

```python
lasso_cv
```

```python
lasso_cv = LassoCV(cv= 5, verbose= 2)
```

```python
lasso_cv
```

```python
lasso_cv.fit(X_train, y_train)
```

```python
lasso_cv
```

```python
y_pred_lasso_cv = lasso_cv.predict(X_test)
r2_lasso_cv = r2_score(y_test, y_pred_lasso_cv)
print(f"The R square of LassoCV regression: {r2_lasso_cv}")
```

```python
from sklearn.linear_model import RidgeCV
ridge_cv = RidgeCV(cv=5)
```

```python
ridge_cv
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
y_pred_ridge_cv = ridge_cv.predict(X_test)
r2_ridge_cv = r2_score(y_test, y_pred_ridge_cv)
print(f"The R square of RidgeCV regression: {r2_ridge_cv}")
```

```python
ridge_cv.get_params()
```

```python
from sklearn.linear_model import ElasticNetCV
elastic_net_cv = ElasticNetCV(cv=5)
```

```python
elastic_net_cv
```

```python
elastic_net_cv.fit(X_train, y_train)
```

```python
y_pred_elastic_cv = elastic_net_cv.predict(X_test)
r2_elastic_cv = r2_score(y_test, y_pred_elastic_cv)
print(f"The R square of ElasticNetCV regression: {r2_elastic_cv}")
```

```python
# Compare all three CV

y_pred_lasso_cv = lasso_cv.predict(X_test)
r2_lasso_cv = r2_score(y_test, y_pred_lasso_cv)
print(f"The R square of LassoCV regression: {r2_lasso_cv}")

y_pred_ridge_cv = ridge_cv.predict(X_test)
r2_ridge_cv = r2_score(y_test, y_pred_ridge_cv)
print(f"The R square of RidgeCV regression: {r2_ridge_cv}")

y_pred_elastic_cv = elastic_net_cv.predict(X_test)
r2_elastic_cv = r2_score(y_test, y_pred_elastic_cv)
print(f"The R square of ElasticNetCV regression: {r2_elastic_cv}")
```
