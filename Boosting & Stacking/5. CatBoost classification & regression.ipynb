# CatBoost Classifier

```python
from catboost import CatBoostClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X, y = make_classification(n_samples= 1000, n_features= 20, n_classes= 2, random_state= 33)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 33)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
classifier = CatBoostClassifier()  #If you don't want details in training put verbose= 0
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

#Evaluate the model
print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print(f"Classification Report: \n{classification_report(y_test, y_pred)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}")
```

```python
# Hyperparameter tuning

params = {
    'n_estimators': [100,200,300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'depth': [3,4,5,6]
}
```

```python
grid_search = GridSearchCV(estimator= CatBoostClassifier(verbose=0), param_grid= params, cv= 5, n_jobs= -1, verbose= 3)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_tuned_pred = best_model.predict(X_test)

#Evaluate the model
print(f"Accuracy Score: {accuracy_score(y_test, y_tuned_pred)}")
print(f"Classification Report: \n{classification_report(y_test, y_tuned_pred)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_tuned_pred)}")
```

# CatBoost Regressor

```python
from catboost import CatBoostRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
X, y = make_regression(n_samples= 1000, n_features= 2, noise= 10, random_state= 35)
X.shape, y.shape
```

```python
regressor = CatBoostRegressor(verbose= 0)
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
# Predict
y_pred = regressor.predict(X_test)

#Evaluate
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
# Parameter Tuning

params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'depth': [3,4,5,6]
}
```

```python
grid_search = GridSearchCV(estimator= CatBoostRegressor(verbose= 0), param_grid= params, cv= 5, n_jobs= -1, verbose= 2)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Best Parameters
print(f"Best Parameters: {grid_search.best_params_}")

# Best Score
print(f"Best Score: {grid_search.best_score_}")

# Best model training
best_model = grid_search.best_estimator_

# Predict
y_tuned_pred = best_model.predict(X_test)

# Evaluation of tuned model
print(f"R2 Score: {r2_score(y_test, y_tuned_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_tuned_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_tuned_pred)}")
```
