#Adaboost classifier

```python
#adaboost classfier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


import warnings
warnings.filterwarnings('ignore')
```

```python
from scipy.sparse import random_array
X,y = make_classification(n_samples= 1000, n_features= 20, n_classes= 2, random_state= 1)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
classifier = AdaBoostClassifier()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Evaluate the model

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0, 1.5, 2.0],
    'algorithm': ['SAMME']
}
```

```python
ada = AdaBoostClassifier()

clf = GridSearchCV(estimator= ada, param_grid= param_grid, cv= 5, verbose= 3)
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
best_model = clf.best_estimator_
best_model
```

```python
y_pred = best_model.predict(X_test)

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print()
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# Adaboost regressor

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X,y = make_regression(n_samples= 1500, n_features= 25, noise= 10)
```

```python
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
regressor = AdaBoostRegressor()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
#Evaluate the model

print("Model Performance")
print()
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
param_grid= {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5, 1.0, 1.5],
    'loss':  ['linear', 'square', 'exponential']
}

reg = GridSearchCV(estimator= regressor, param_grid= param_grid, cv= 5, verbose= 3)
reg
```

```python
reg.fit(X_train, y_train)
```

```python
reg.best_params_
```

```python
reg.best_score_
```

```python
best_model = reg.best_estimator_
best_model
```

```python
y_pred_tuned = best_model.predict(X_test)
```

```python
print("Regressor Model Performance")
print()
print(f"R2 Score: {r2_score(y_test, y_pred_tuned)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred_tuned)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_tuned)}")
```
