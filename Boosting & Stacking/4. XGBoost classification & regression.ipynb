## XG Boost Regressor

```python
# XGBoost Regressor

from xgboost import XGBRegressor
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X, y = make_regression(n_samples= 1000, n_features= 2, noise= 10, random_state=12)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
regressor = XGBRegressor()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)

#Evaluate the model
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.5, 0.05, 0.1, 0.2]
}
```

```python
gs_reg = XGBRegressor()
grid_search = GridSearchCV(estimator= gs_reg, param_grid= param_grid, cv= 5, verbose= 3, n_jobs= -1)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_

y_tuned_pred = best_model.predict(X_test)

#Evaluate the model
print(f"R2 Score: {r2_score(y_test, y_tuned_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_tuned_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_tuned_pred)}")
```

## XG Boost Classifier

```python
from xgboost import XGBClassifier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Generate dataset
X, y = make_classification(n_samples= 1000, n_features= 20, n_classes= 2, random_state=42)
```

```python
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf = XGBClassifier()
clf
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

#Evaluate the model
print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.5, 0.05, 0.1, 0.2]
}
```

```python
gs_clf = XGBClassifier()
grid_search = GridSearchCV(estimator= gs_clf, param_grid= param_grid, cv= 5, verbose= 3, n_jobs= -1)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_tuned_pred = best_model.predict(X_test)

#Evaluate the model
print(f"Accuracy Score: {accuracy_score(y_test, y_tuned_pred)}")
print(f"Classification Report:\n{classification_report(y_test, y_tuned_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_tuned_pred)}")
```
