```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_regression(n_samples= 1200, n_features= 3, noise= 10)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
regressor = GradientBoostingRegressor()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Evaluate the model

print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Param grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.05, 0.2],
#     'max_depth': [3, 4, 5],
#     'subsample': [0.8, 0.9, 1.0],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4]
}
```

```python
gb_regressor = GradientBoostingRegressor()

grid_search = GridSearchCV(estimator= gb_regressor, param_grid= param_grid, cv= 5, verbose= 3)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_tuned_pred = best_model.predict(X_test)

# Evaluate Tuned model
print(f"R2 Score: {r2_score(y_test, y_tuned_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_tuned_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_tuned_pred)}")
```
