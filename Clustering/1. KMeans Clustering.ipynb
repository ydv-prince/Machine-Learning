```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples= 1000, centers= 3, n_features= 2, random_state= 44)
X.shape, y.shape
```

```python
y
```

```python
# In Unsupervised learning you will not use y

X
```

```python
plt.scatter(X[:,0], X[:, 1], c= y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 44)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= 3)
kmeans
```

```python
kmeans.fit(X_train)   # kmeans.fit_predict(X_train) can also be used to get the prediction
```

```python
y_labels = kmeans.predict(X_train)
```

```python
y_labels
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_labels)
```

```python
# Evaluation: How better this clustering model is ?

from sklearn.metrics import silhouette_score
silhouette_score(X_train, kmeans.labels_) #close to 1, we have a good model
```

```python
kmeans.labels_
```

```python
kmeans.inertia_  # k=3, this is wcss distance
```

```python
#how to select k(centroid) >> elbow method
```

```python
wcss = []
for k in range(1, 10):
  kmeans = KMeans(n_clusters= k)
  kmeans.fit(X_train)
  wcss.append(kmeans.inertia_)
```

```python
wcss  # Here, wcss value is decreasing as k inceasing
```

```python
#plot the elbow curve

plt.plot(range(1, 10), wcss)
plt.xticks(range(1, 10))
plt.xlabel("No of students")
plt.ylabel("WCSS")
plt.show()
```

```python
#based on the plot above the optimal no of k/centroid should be 3
```

```python
# Lets see by elbow method

# !pip install kneed

from kneed import KneeLocator

k1 = KneeLocator(range(1,10), wcss, curve= 'convex', direction= 'decreasing')
```

```python
k1.elbow
```

```python
silhouette_coeff = []
for k in range(2,10):
  kmeans = KMeans(n_clusters= k)
  kmeans.fit(X_train)
  score = silhouette_score(X_train, kmeans.labels_)
  silhouette_coeff.append(score)
```
